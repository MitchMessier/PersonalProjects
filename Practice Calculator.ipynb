```python
def add(x,y):
    return x+y

def subtraction(x,y):
    return x-y

def multiplication(x, y):
    return x*y

def division(x, y):
  if y == 0:
    return "Error! Division by zero."
  else:
    return x/y
```

```python
def menu():
  print("select operation:")
  print("1. addition")
  print("2. subtraction")
  print("3. multiplication")
  print("4. division")
  print("5. exit")
```

```python
def user_input():
  while True:
    menu()
    choice = input("Enter choice: (1/2/3/4/5): ")

    if choice == '5':
      print("Exiting the calculator. Goodbye!")
      break

    if choice in ['1', '2', '3', '4']:
      try:
        num1 = float(input('Enter your first number:'))
        num2 = float(input('Enter your secound number:'))
      except ValueError:
        print("This is not a valid number. Please enter a valid number")
        continue

        if choice == '1':
          print(f'{num1} + {num2} = {add(num1, num2)}')
        elif choice == '2':
          print(f'{num1} + {num2} = {add(num1, num2)}')
        elif choice == '3':
          print(f'{num1} + {num2} = {add(num1, num2)}')
        elif choice == '4':
          print(f'{num1} + {num2} = {add(num1, num2)}')

        else:
          print('Not a valid number')
```
